
## Higher order functions

As we learned from `sort/filter/map`, sometimes we want to use a function as an argument to another function.
E.g. `sort` takes a function as an argument (which tells sort how to order the items in a list). `Filter` takes a function
as an argument (which tells filter how to decide which items pass through). And `map` takes a function as an argument
(which tells map how to transform elements of the old list into elements in the new list).

Can we use this technique in our own functions? Here is a function that reads a FASTA file and returns a dict where the keys are the names and the values are the sequences.

Remember that we can take a 'slice' of a string (or List) by saying

    my_string[start:stop]

e.g. this will get the third, fourth and fifth characters in `my_string`

    my_string[3:6]

If we leave off the `stop` number, we get all characters until the end of the string. So this:

    my_string[1:]

is a handy shorthand for all of `my_string` except the first character. 


```python
def parse_fasta(filename):
    f = open(filename)
    sequences = {}
    for line in f:
        if line.startswith('>'):
            name = line[1:].rstrip('\n')
            sequences[name] = ''
        else:
            sequences[name] = sequences[name] + line.rstrip('\n')
    return sequences
```

```python
print(parse_fasta('big.fasta'))
```

In a way, this function is the opposite of the `format_fasta` function we were working with before. `parse_fasta` splits FASTA sequences up, `format_fasta` puts them back together. Of course, it's important to notice that `format_fasta` works on mulitple sequences, whereas `parse_fasta` only works on one sequence at a time. 

If we put the two functions together, we can write a new function that reads FASTA sequence from a file and splits them into a dict (using `parse_fasta`) then loops through the sequences and writes them to an output file in FASTA format (using `format_fasta`). We will call this new function `fasta_copy` since that is what it does.

```python
# remember our old friend format_fasta?
def format_fasta(name, sequence):
    fasta_string = '>' + name + "\n" + sequence + "\n"
    return fasta_string

# we can use parse_fasta and format_fasta together
for name, sequence in parse_fasta('big.fasta').items():
    print(format_fasta(name, sequence))
    
# here is a very boring function that copies FASTA sequences
# from one file to another (by parsing them)
def fasta_copy(input_filename, output_filename):
    output_file = open(output_filename, 'wt')
    for name, sequence in parse_fasta(input_filename).items():
        output_file.write(format_fasta(name, sequence))

fasta_copy('big.fasta', 'copy.fasta')    
    
```

Here are a couple of variations on `fasta_copy`. 

The first one removes the 
first 4 bases of each sequence before writing it to the output file. 


The second one removes the first 8 bases. We could use these to e.g. remove adapter sequences
from the start of sequencing reads. 

```python
# a function that removes the first 4 bases of each sequence
def fasta_trim_4(input_filename, output_filename):
    output_file = open(output_filename, 'wt')
    for name, sequence in parse_fasta(input_filename).items():
        new_sequence = sequence[4:]
        output_file.write(format_fasta(name, new_sequence))

fasta_trim_4('big.fasta', 'trim4.fasta')

# a function that removes the first 8 bases of each sequence
def fasta_trim_8(input_filename, output_filename):
    output_file = open(output_filename, 'wt')
    for name, sequence in parse_fasta(input_filename).items():
        new_sequence = sequence[8:]
        output_file.write(format_fasta(name, new_sequence))

fasta_trim_8('big.fasta', 'trim8.fasta')
```

These functions are nearly identical - let's make a generic version that will take
the number of bases to be trimmed as an argument.

```python


def fasta_trim_any(input_filename, output_filename, trim):
    output_file = open(output_filename, 'wt')
    for name, sequence in parse_fasta(input_filename).items():
        new_sequence = sequence[trim:]
        output_file.write(format_fasta(name, new_sequence))

fasta_trim_any('big.fasta', 'trim4.fasta', 4)
fasta_trim_any('big.fasta', 'trim8.fasta', 8)
```

What we have done in creating `fasta_trim_any` is make a more generic version of `fasta_trim_4` that can handle trimming any number of bases.

Here is a very similar function that converts the DNA sequences to lower
case before writing them to the output file. Notice how much of the code is shared
between this function and `trim_any` in the box above.

```python


def fasta_to_lower(input_filename, output_filename):
    output_file = open(output_filename, 'wt')
    for name, sequence in parse_fasta(input_filename).items():
        new_sequence = sequence.lower()
        output_file.write(format_fasta(name, new_sequence))
        
fasta_to_lower('big.fasta', 'lower.fasta')
```

Just like we made a more generic version of `trim4` that could handle any number
of bases, can we do something similar and make an even more generic version that
can also handle other type of modifications of the dna sequence?

We can try using default arguments. We'll have our function take a `trim` argument, which is the number of bases to remove from the start (default: zero) and a `lower` argument, which is whether or not to convert the DNA sequence to lower case (default: false).

```python

def fasta_trim_any_and_lower(input_filename, output_filename, trim=0, lower=False):
    output_file = open(output_filename, 'wt')
    for name, sequence in parse_fasta(input_filename).items():
        new_sequence = sequence[trim:]
        if (lower):
            new_sequence = new_sequence.lower()
        output_file.write(format_fasta(name, new_sequence))

fasta_trim_any_and_lower('big.fasta', 'trim8.fasta', trim=8)
fasta_trim_any_and_lower('big.fasta', 'lower.fasta', lower=True)
```

This works for the examples we are looking at, but it's not very satisfying - if we want to add another type of transformation to the DNA sequence (say, truncate it to a maximum length) then we will need to add yet another argument. Here is a much better way to do it - we will write our function so that it takes as one its arguments the name of another function which it will run on each DNA sequence before writing it to the output file.

Now we can write our `trim_4`, `trim_8` and `lower` functions separately, and `process_fasta` will change its behaviour depending on which one we give as an argument.

```python
# here's a better way to do it
# we make the thing that we want to do to each sequence an argument itself

def process_fasta(input_filename, output_filename, the_function):
    output_file = open(output_filename, 'wt')
    for name, sequence in parse_fasta(input_filename).items():
        new_sequence = the_function(sequence)
        output_file.write(format_fasta(name, new_sequence))

# now we can write everything much more succinctly

def lower_function(input):
    return input.lower()

def trim_4_function(input):
    return input[4:]

def trim_8_function(input):
    return input[8:]


process_fasta('big.fasta', 'lower.fasta', lower_function)
process_fasta('big.fasta', 'trim4.fasta', trim_4_function)
process_fasta('big.fasta', 'trim8.fasta', trim_8_function)

```

This may well remind you of how we use the built-in Python `map` function.  Just as `map()` takes as one of its inputs a function describing how to modify 
list elements, `process_fasta()` takes as one of its inputs a function describing 
how to modify dna sequences. `process_fasta()` is really a very specialized version
of `map()` that works on FASTA files instead of lists. 

A few more examples....

```python
# now we can process FASTA files using arbitrary functions

# this one replaces all EcoRI sites with X's
def mask_ecori_sites(input):
    return input.replace('GAATTC', 'XXXXXX')

process_fasta('big.fasta', 'masked.fasta', mask_ecori_sites)

# we can use any function that has a string argument and a string return value
# here is one from previously
def ambig_to_lower(input):
    output = ''
    for base in input.upper():
        if base in ['A', 'T', 'G', 'C']:
            output = output + base
        else:
            output = output + base.lower()
    return output

process_fasta('big.fasta', 'ambig_lowered.fasta', ambig_to_lower)
```

```python
# with lambda functions, we can be even more concise

process_fasta('big.fasta', 'lambda_lower.fasta', lambda x : x.lower())
process_fasta('big.fasta', 'lambda_trim4.fasta', lambda x : x[4:])
process_fasta('big.fasta', 'lambda_trim8.fasta', lambda x : x[8:])
```

We have removed much of the repetition from our code; however, there is still some
left - look at our two trim functions:

    def trim_4_function(input):
        return input[4:]

    def trim_8_function(input):
        return input[8:]

What we need is a kind of function factory - a function who's job is to build
other functions. We will give this factory a number, and it will give us back
a new function that removes that number of bases from the begining of a string.

```python


# a function to build functions that trim sequences
def build_trim_function(trim):
    def new_function(input):
        return input[trim:]
    return new_function

trim4 = build_trim_function(4)
trim8 = build_trim_function(8)

print(trim4('abcdefghijklmnopqrstuvwxyz'))
print(trim8('abcdefghijklmnopqrstuvwxyz'))

process_fasta('big.fasta', 'trim4.fasta', trim4)
process_fasta('big.fasta', 'trim8.fasta', trim8)

process_fasta('big.fasta', 'trim4.fasta', build_trim_function(4))
process_fasta('big.fasta', 'trim8.fasta', build_trim_function(8))
```

This is a very powerful idea - rather than writing a definition of a function every
time we need one, we can use our factory to build them. It only works for certain
cases though!

These two techniques are collectively called **Higher order programming** and can be used to solve some problems in a very elegant way.

## Recursive functions

The Newick tree format lets us write phylogenetic trees using brackets. e.g. this (taken from http://evolution.genetics.washington.edu/phylip/newicktree.html)

      |--------------dog
      |
      |          |---raccoon
      |----------|  
      |          |---bear
      |
      |          |---sea lion
      |   |------|  
      |   |      |---seal
      |   |
      |---|      |---monkey
          |   |--|  
          |   |  |---cat
          |---|
              |------weasel

is represented as 

    (dog,(raccoon,bear),((sea_lion,seal),((monkey,cat),weasel)));

An interesting fact is that if we change the () to [] and put quotes around the names we get a valid bit of python code

    ['dog',['raccoon','bear'],[['sea_lion','seal'],[['monkey','cat'],'weasel']]]

Which forms a representation of the tree as a nested set of lists.

```python
tree = ['dog',['raccoon','bear'],[['sea_lion','seal'],[['monkey','cat'],'weasel']]]
for node in tree:
    print(node)
```

Suppose we want to print a nicely-indented version of this tree which shows for each node the number of branches away from the root. We know that each element of the list can be either a string (the name of one of the leaf nodes) or another list, so we will start by writing an `if` statement to deal with these two possibilities.  We can test what 'kind' of thing a variable is in Python by using the `isinstance()` function. Here, we want to check each element in a list to see if it is a string, so we can say:

    if isinstance(some_variable, str)


```python
for node in tree:
    # if the node is a leaf, i.e. a string, then print it
    if isinstance(node, str):
        print(node)
    # otherwise, go down one level and print it with some spaces
    else:
        for lower_node in node:
            print('  ' + str(lower_node))
```

This works for two levels but there are more than two levels in our nested list, so let's add some more code to deal with the next level:

```python
for node1 in tree:
    if isinstance(node1, str):
        print(node1)
    else:
        for node2 in node1:
            if isinstance(node2, str):
                print('  ' + str(node2))
            else:
                print('    ' + str(node2))
```

Now the first two levels are find but we are still not printing the third level right. More code:

```python
for node1 in tree:
    if isinstance(node1, str):
        print(node1)
    else:
        for node2 in node1:
            if isinstance(node2, str):
                print('  ' + str(node2))
            else:
                for node3 in node2:
                    if isinstance(node3, str):
                        print('    ' + str(node3))
                    else:
                        print('      ' + str(node3))
```

This approach is clearly not going to work - the code is becoming very repetitive and hard to read and it seems like we have to know in advance how many levels there are in our list. We can move some of the formatting code into a separate function:

```python

def print_node(node, level):
        print(('  ' * level) + str(node))

level = 0
for node1 in tree:
    if isinstance(node1, str):
        print_node(node1, level)
    else:
        for node2 in node1:
            if isinstance(node2, str):
                print_node(node2, level+1)
            else:
                for node3 in node2:
                    if isinstance(node3, str):
                        print_node(node3, level+2)
                    else:
                        print_node(node3, level+3)
```

But that doesn't really improve matters. 

What we really want to do is run the same code at each level. We know how to print a node that is a string, and in order to print a node that is a list, we just need to do the same thing for each element.

On other words, to print a node, we want to do the following: if the node is a string, then just print it (with some number of spaces in front). If it's a list, then print each node in the list. 

Here is that idea as code:

```python

def print_node(node, level):
    if isinstance(node, str):
        print(('  ' * level) + str(node))
    else:
        for child in node:
            # here is the magic
            # call print_node recursively
            print_node(child, level+1)
            
print_node(tree, 0)
```

```python
# now it works on any size input
# tree taken from http://phylogram.org/
big_tree = ['Edentata', [[['Orycteropus', 'Trichechus'], ['Procavia', ['Elephas', 'Loxodonta']]], [[['Chiroptera', ['Tupaia', [['Talpa', ['Suncus', 'Erinaceus']], ['Manis', [['Felis', 'Leo'], [['Canis', 'Ursidae'], [['Phocidae', 'Zalophus'], ['Procyonidae', 'Mustelidae']]]]]]]], [[['Lemuridae', ['Galago', 'Nycticebus']], ['Tarsius', [['Cebus', ['Atelinae', ['Aotus', 'Callithrix']]], [['Hylobates', ['Pongo', ['Gorilla', ['Homo', 'Pan']]]], ['Presbytis', ['Cercopithecus', ['Erythrocebus', ['Macaca_fascicularis', ['Macaca_mulatta', 'Macaca_fuscata'], ['Theropithecus', 'Papio']]]]]]]]], [['Ochotona', 'Oryctolagus'], ['Caviomorpha', ['Spermophilus', ['Spalax', [['Rattus', 'Mus'], ['Ondatra', 'Mesocricetus']]]]]]]], [['Sus', [['Lama', 'Camelus'], ['Hippopotamus', [[['Ovis', 'Capra'], [['Antilocapra', 'Giraffa'], ['Cervus', 'Alces']]], ['Tragelaphinae', ['Bos_grunniens', ['Bison', 'Bos_taurus']]]]]]], [[['Equus_Asinus', 'Equus' 'caballus'], ['Tapirus', 'Rhinocerotidae']], [['Phocoena', 'Tursiops'], ['Balaenoptera', 'Eschrichtius']]]]]]]
print_node(big_tree, 0)
```

This is known as a **recursive function** and it works will for tree-like data structures.
